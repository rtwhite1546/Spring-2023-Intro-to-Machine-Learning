### Homework 1 (35 points)

Ridge regression is am modified version of linear regression that penelizes the coefficients for being large. It accomplishes this by adding a so-called $L^2$ penalty term to the loss function (e.g. mean squared error):

$$L(\theta)=\frac{1}{n}\sum\limits_{i=1}^n \left(\hat{f}(x_i)-y_i\right)^2 + \lambda\sum\limits_{i=1}^d \theta_i^2$$

where $\lambda>0$ is a **hyperparameter** that must be tuned by the user. An appropriate choice of $\lambda$ can often help with learning datasets where the input features are highly correlated or it can help with an overfitting problem.

(a) Write each part of $L(\theta)$ in matrix-vector form where $\hat{f}$ is a LBF expansion regression model. Define each matrix and vector separately by writing their elements with subscripts, and state their dimensions.

(b) Solve the following optimization problem by hand for the loss function $L$ above:

$$\min\limits_{\theta\in\mathbb{R}^{d+1}} L(\theta)$$

(c) Write a Python class for this ridge LBF expansion regression model with a `fit` function applying the formula from part (b) to compute the parameters $\theta$ and a `predict` function to make predictions for input data after the model has been fit.

(d) Apply this class to fit a LBF regression model to predicting real estate prices based on the 6 input features in the [Real estate valuation dataset on the UCI Machine Learning Database](https://archive.ics.uci.edu/ml/datasets/Real+estate+valuation+data+set).

(e) Test at least 3 different LBF expansions (best-fit hyperplane, best-fit quadratic surface, and an alternative one of your choice), each with at least 5 different values of the hypereparameter $\lambda$. (You should have at least 15 experiments.)

Document all experiments and conclude which experiment works best based on an appropriate train/test split. Report your best test-set MAE at the end of your submission.
